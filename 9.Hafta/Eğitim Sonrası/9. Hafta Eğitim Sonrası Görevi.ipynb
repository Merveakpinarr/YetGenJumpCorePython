# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import random
import time

class Kumanda:
    def __init__(self):
        self.tv_durumu = "Kapalı"
        self.ses = 0
        self.kanal_listesi = []
        self.aktif_kanal = ""

    def tv_ac(self):
        self.tv_durumu = "Açık"
        print("Televizyon açıldı.")

    def tv_kapat(self):
        self.tv_durumu = "Kapalı"
        print("Televizyon kapatıldı.")

    def ses_ayarları(self):
        yeni_ses = int(input("Yeni ses seviyesini girin: "))
        self.ses = yeni_ses
        print(f"Ses ayarlandı: {self.ses}")

    def kanal_ekle(self):
        kanal = input("Eklemek istediğiniz kanalı girin: ")
        self.kanal_listesi.append(kanal)
        print(f"{kanal} kanalı eklendi.")

    def kanal_sayısını_öğrenme(self):
        kanal_sayisi = len(self.kanal_listesi)
        print(f"Toplam kanal sayısı: {kanal_sayisi}")

    def rastgele_kanala_geçme(self):
        if len(self.kanal_listesi) > 0:
            self.aktif_kanal = random.choice(self.kanal_listesi)
            print(f"Aktif kanal: {self.aktif_kanal}")
        else:
            print("Kanal bulunmamaktadır.")

    def televizyon_bilgileri(self):
        print("Televizyon Durumu:", self.tv_durumu)
        print("Ses Seviyesi:", self.ses)
        print("Aktif Kanal:", self.aktif_kanal)

    def kumandayı_kapat(self):
        print("Kumanda kapatılıyor...")
        time.sleep(1)
        print("Kumanda kapatıldı.")

    def kumandayı_başlat(self):
        while True:
            print("""
            Kumanda İşlemleri:
            1. Tv Aç
            2. Tv Kapat
            3. Ses Ayarları
            4. Kanal Ekle
            5. Kanal Sayısını Öğrenme
            6. Rastgele Kanala Geçme
            7. Televizyon Bilgileri
            q. Kumandayı Kapat
            """)

            islem = input("İşlem seçin: ")

            if islem == "1":
                self.tv_ac()
            elif islem == "2":
                self.tv_kapat()
            elif islem == "3":
                self.ses_ayarları()
            elif islem == "4":
                self.kanal_ekle()
            elif islem == "5":
                self.kanal_sayısını_öğrenme()
            elif islem == "6":
                self.rastgele_kanala_geçme()
            elif islem == "7":
                self.televizyon_bilgileri()
            elif islem.lower() == "q":
                self.kumandayı_kapat()
                break
            else:
                print("Geçersiz işlem. Tekrar deneyiniz")


```
