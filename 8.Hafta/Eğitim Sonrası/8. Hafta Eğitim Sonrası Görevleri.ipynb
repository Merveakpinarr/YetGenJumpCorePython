# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız
def calculate_bmi(weight, height):
    height_in_meters = height / 100
    bmi = weight / (height_in_meters ** 2)
    return bmi


def determine_bmi_category(bmi):
    if bmi < 18.5:
        return "Zayıf"
    elif bmi < 25:
        return "Normal kilolu"
    elif bmi < 30:
        return "Fazla kilolu"
    elif bmi < 40:
        return "Obez"
    else:
        return "İleri derecede obez"


def main():
    entry_count = int(input("Kaç giriş yapmak istiyorsunuz? "))

    entries = []
    for i in range(entry_count):
        print(f"\n--- Giriş {i + 1} ---")
        name_surname = input("Ad ve soyad: ")
        age = int(input("Yaş: "))
        height = int(input("Boy (cm): "))
        weight = float(input("Kilo (kg): "))

        bmi = calculate_bmi(weight, height)
        category = determine_bmi_category(bmi)

        entry = {
            "Ad Soyad": name_surname,
            "Yaş": age,
            "Boy": height,
            "Kilo": weight,
            "BMI": bmi,
            "Kategori": category
        }
        entries.append(entry)

    
    with open("veriler.txt", "w") as file:
        for entry in entries:
            file.write(f"Ad Soyad: {entry['Ad Soyad']}\n")
            file.write(f"Yaş: {entry['Yaş']}\n")
            file.write(f"Boy: {entry['Boy']}\n")
            file.write(f"Kilo: {entry['Kilo']}\n")
            file.write(f"BMI: {entry['BMI']}\n")
            file.write(f"Kategori: {entry['Kategori']}\n")
            file.write("\n")

    print("\nVeriler başarıyla kaydedildi.")


main()
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python

# 2.Problemin Çözümünü Buraya Yazınız
import datetime

class SmoothieBank:
    def __init__(self):
        self.menu = {
            "1": {"name": "Detox", "ücret": 10},
            "2": {"name": "Enerji", "ücret": 12},
            "3": {"name": "Protein", "ücret": 15},
            "4": {"name": "Meyve Karışımlı", "ücret": 8},
            "5": {"name": "Yeşil Smoothie", "ücret": 10}
        }
        self.transactions = []

    def display_menu(self):
        print("--- Smoothie Bank Menü ---")
        for key, value in self.menu.items():
            print(f"{key}. {value['name']} - {value['price']} TL")
        print()

    def purchase(self):
        choice = input("Lütfen satın almak istediğiniz smoothie numarasını girin (Çıkış için 'q'): ")
        if choice == 'q':
            return

        if choice in self.menu:
            smoothie = self.menu[choice]
            timestamp = datetime.datetime.now()
            transaction = {"smoothie": smoothie["name"], "price": smoothie["price"], "timestamp": timestamp}
            self.transactions.append(transaction)
            print(f"\n{smoothie['name']} smoothie satın aldınız. Ödeme: {smoothie['price']} TL\n")
        else:
            print("Geçersiz bir seçim yaptınız.\n")

    def show_transactions(self):
        print("--- Satın Alımlar ---")
        for transaction in self.transactions:
            timestamp = transaction["timestamp"].strftime("%Y-%m-%d %H:%M:%S")
            smoothie = transaction["smoothie"]
            price = transaction["price"]
            print(f"Tarih: {timestamp}, Smoothie: {smoothie}, Ödeme: {price} TL")
        print()

def main():
    smoothie_bank = SmoothieBank()

    while True:
        print("YETBAR - SMOOTHIE BANK\n")
        print("1. Menüyü Görüntüle")
        print("2. Satın Alma")
        print("3. Satın Alımları Göster")
        print("4. Çıkış")

        choice = input("Lütfen bir işlem seçin: ")

        if choice == '1':
            smoothie_bank.display_menu()
        elif choice == '2':
            smoothie_bank.purchase()
        elif choice == '3':
            smoothie_bank.show_transactions()
        elif choice == '4':
            print("Programdan çıkılıyor...")
            break
        else:
            print("Geçersiz bir seçim yaptınız. Tekrar deneyin.\n")

if __name__ == "__main__":
    main()
```
