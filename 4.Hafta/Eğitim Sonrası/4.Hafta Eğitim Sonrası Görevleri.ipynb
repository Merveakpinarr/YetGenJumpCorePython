# Problem 2

### a - Biraz Ä±sÄ±narak baÅŸlayalÄ±m!

```
numbers = [5,10,10,15,15,15,20,20,20,20]
```

YukarÄ±da `numbers` adÄ±nda 10 elemanlÄ± bir liste verilmiÅŸtir. Bu listenin her elemanÄ±nÄ± kontrol eden ve tekrar eden elemanlarÄ± bir kez olacak ÅŸekilde baÅŸka bir listeye kopyalayan bir program yazÄ±nÄ±z. Son olarak oluÅŸturduÄŸunuz listeyi ekrana bastÄ±rÄ±nÄ±z.




```
>Output
[5, 10, 15, 20]
```


````python
numbers = [5, 10, 10, 15, 15, 15, 20, 20, 20, 20]

unique_numbers = []

for number in numbers:
    if number not in unique_numbers:
        unique_numbers.append(number)

print(unique_numbers)
````

### ğŸŒ¶ï¸ b) Hadi iÅŸi biraz daha zorlaÅŸtÄ±ralÄ±m! 


Åimdiyse senden `numbers` isimli listedeki tekrar eden elemanlarÄ± `repeatNumbers` adÄ±ndaki listeye; tekrar etmeyen elemanlarÄ± `uniqueNumbers`adÄ±ndaki listeye kopyalayan yazmanÄ± ve listeleri ekrana bastÄ±rmanÄ± bekliyoruz

```
>Output
Tekrar Eden Elemanlar: [10, 10, 15, 15, 15, 20, 20, 20, 20]
Tekrar Etmeyen Elemanlar: [5]
```

````python
numbers = [5, 10, 10, 15, 15, 15, 20, 20, 20, 20]

repeatNumbers = []
uniqueNumbers = []

for number in numbers:
    if number in uniqueNumbers:
        repeatNumbers.append(number)
    else:
        uniqueNumbers.append(number)

print("Tekrar Eden Elemanlar:", repeatNumbers)
print("Tekrar Etmeyen Elemanlar:", uniqueNumbers)
````
